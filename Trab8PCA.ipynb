```python
from sklearn.datasets import fetch_mldata
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.model_selection import train_test_split
import pandas as pd


#mnist = fetch_openml('mnist_784')
mnist_data = fetch_openml('mnist_784', version=1)
X, y = mnist_data['data'], mnist_data['target']

#mnist


#mnist.data.shape


#mnist.target.shape


'''train_img, test_img, train_lbl, test_lbl = train_test_split(
    mnist.data, mnist.target, test_size=1/7.0, random_state=0)'''

train_img, test_img, train_lbl, test_lbl = X[:60000], X[60000:], y[:60000], y[60000:]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


scaler.fit(train_img)


train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
train_regr = train_img
test_regr = test_img

from sklearn.decomposition import PCA

pca = PCA(.95)

pca.fit(train_img)

pca.n_components_

train_img = pca.transform(train_img)
test_img = pca.transform(test_img)

from sklearn.linear_model import LogisticRegression


logisticRegr = LogisticRegression(solver = 'lbfgs')
logisticRegr2 = LogisticRegression(solver = 'lbfgs')


logisticRegr.fit(train_img, train_lbl)
logisticRegr2.fit(train_regr, train_lbl)

logisticRegr.predict(test_img[0].reshape(1,-1))
logisticRegr2.predict(test_regr[0].reshape(1,-1))


logisticRegr.predict(test_img[0:10])
logisticRegr2.predict(test_regr[0:10])

score = logisticRegr.score(test_img, test_lbl)
score2 = logisticRegr2.score(test_regr, test_lbl)

print('Com PCA:' , score , '\n' , 'Sem PCA: ' , score2 , '\n')

```
